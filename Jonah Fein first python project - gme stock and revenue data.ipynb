```python
import yfinance as yf
import pandas as pd
import requests
from bs4 import BeautifulSoup
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

```python
gme = yf.Ticker("GME")
gme_data = gme.history(period = 'max')
gme_data
```

```python
gme_data.reset_index(inplace = True)
# Convert 'Date' column to datetime in gme_data
gme_data['Date'] = pd.to_datetime(gme_data['Date'])
gme_data.head()
```

```python
import requests

url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/stock.html"

html_data = requests.get(url).text
```

```python
soup = BeautifulSoup(html_data, )
read_html_pandas_data = pd.read_html(url)
read_html_pandas_data = pd.read_html(str(soup))
gme_dataframe = read_html_pandas_data[0]
gme_dataframe
```

```python
gme_dataframe.columns = ['Date', 'Revenue']
gme_dataframe['Revenue'] = gme_dataframe['Revenue'].str.replace(',','').str.replace('$','')
gme_dataframe['Revenue'] = pd.to_numeric(gme_dataframe['Revenue'], errors='coerce')
gme_dataframe
```

```python
#test to make sure it is numeric
print('Test number:', gme_dataframe.loc[0, 'Revenue'])
print('Test number type:', type(gme_dataframe.loc[0,'Revenue']))
```

```python
gme_dataframe.tail()
```

```python
gme_dataframe['Date'] = pd.to_datetime(gme_dataframe['Date'], errors='coerce')
gme_dataframe_for_graph = gme_dataframe.loc[0:10000, ['Date','Revenue']]
gme_dataframe_for_graph
```

```python
gme_dataframe_for_graph.plot(x = 'Date', y = 'Revenue', title='GME Revenue', xlabel='Date', ylabel='Revenue')
```
